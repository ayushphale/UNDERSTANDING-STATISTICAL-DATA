```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
df=pd.DataFrame({'Category':['A','B','C','D','E'],
                 'Values':[50,7,30,90,60],
                 'Sales':np.random.randint(100,500,5),
                 'Profit':np.random.randint(10,100,5)})
print(df)
```

```python
#Area Plot
plt.figure(figsize=(10,6))
plt.fill_between(df['Category'],df['Values'],color='blue',alpha=0.4)
plt.title('Area Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
#Pie Chart
plt.figure(figsize=(10,6))
plt.pie(df['Values'],labels=df['Category'],autopct='%1.2f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Donut Chart
plt.figure(figsize=(10,6))
plt.pie(df['Values'],labels=df['Category'],autopct='%1.1f%%')
centre_circle=plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Donut Chart')
plt.show()
```

```python
# Boxplot (outliers Detection)
plt.figure()
sns.boxplot(x=df['Values'])
plt.title('Boxplot of sales')
plt.show()
```

```python
#Heatmap(Correlations)
plt.figure()
corr=df[['Sales','Profit']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Bubble Plot
plt.scatter(df['Sales'],df['Profit'],s=df['Values']*10)
plt.title('Bubble Plot')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales',y='Profit',
               size='Values', #bubble size
               hue='Values',  #colour
               data=df,
               sizes=(50,300),
               palette='viridis')
plt.title("Bubble Plot using Seaborn")
plt.show()
```
